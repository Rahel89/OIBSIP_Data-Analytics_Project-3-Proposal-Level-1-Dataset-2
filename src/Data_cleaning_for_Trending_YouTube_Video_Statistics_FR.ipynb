```python
# Converting JSON to a DataFrame
categories_df = pd.DataFrame(FR_json["items"])

categories_df = pd.DataFrame(list(categories_df['items']))
categories_df = categories_df[["id", "snippet"]].copy()
categories_df["category_name"] = categories_df["snippet"].apply(lambda x: x["title"])
categories_df = categories_df[["id", "category_name"]].copy()
categories_df.rename(columns={"id": "category_id"}, inplace=True)
categories_df["category_id"] = categories_df["category_id"].astype(str)


if 'category_id' not in FR.columns:
    if 'id' in FR.columns:
        FR.rename(columns={'id': 'category_id'}, inplace=True)

    else:
        print(f"Error: Column named 'category_id' or is missing in the FR DataFrame")
        print(f"Existing FR columns: {FR.columns.to_list()}")

FR["category_id"] = FR["category_id"].astype(str)

FR = FR.merge(categories_df, on='category_id', how='left')
FR.drop(columns=['category_id'], inplace=True)
print(FR.head())
```

#Outlier detection

The method chosen to identify outliers is the standard deviation method. After identifying, the IQR method will be employed to addressed the outliers because it handles outliers based on the spread of the data. After addressing the outliers, a box plot will be used to visualize the final result.

threshold = 3

**1. views**

```python
threshold = 3
mean = FR['views'].mean()
std = FR['views'].std()

FR['outlier_flag'] = (FR['views'] - mean).abs() > threshold * std

outliers = FR[FR['outlier_flag']]
print("Outliers:")
print(outliers)
```

```python
Q1 = FR['views'].quantile(0.25)
Q3 = FR['views'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = FR[(FR['views'] >= lower_bound) & (FR['views'] <= upper_bound)]
```

```python
#distribution of the data before and after outlier removal
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=FR['views'], color ='brown')
plt.title('Box Plot before Outlier Removal')

plt.subplot(1, 2, 2)
sns.boxplot(y=df_cleaned['views'], color='brown')
plt.title('Box Plot after Outlier Removal')
```
